```python
print(DoS_data["CANID"].nunique())
print(Fuzzy_data["CANID"].nunique())
print(RPM_data["CANID"].nunique())
```

```python
def get_interval_id(ds):
  dict_interval_id = {}
  interval_ids = []
  for idx, data in ds.iterrows():
    if data["CANID"] in dict_interval_id:
      tmp = data["Timestamp"] - dict_interval_id[data["CANID"]]

    else:
      tmp = np.NaN
    interval_ids.append(tmp)
    dict_interval_id[data["CANID"]] = data["Timestamp"]
  #ds['interval_id'] = interval_ids
  return interval_ids

tmp_dos = get_interval_id(DoS_data)
DoS_data['interval_id'] = tmp_dos

tmp_fuzzy = get_interval_id(Fuzzy_data)
Fuzzy_data['interval_id'] = tmp_fuzzy

tmp_rpm = get_interval_id(RPM_data)
RPM_data['interval_id'] = tmp_rpm
```
